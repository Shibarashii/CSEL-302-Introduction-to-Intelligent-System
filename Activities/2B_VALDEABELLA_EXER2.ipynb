# **I .Data Loading**

```python
import pandas as pd
import matplotlib.pyplot as plt

dataFrame = pd.read_csv('hardwareStore.csv')
```

# **II. Initial Data Analysis**

```python
basicStat = dataFrame.describe()
uniqueCategories = dataFrame['CATEGORY_NAME'].value_counts()
```

# **III. Category Analysis**

```python
plt.figure(figsize=(10, 6))
dataFrame['CATEGORY_NAME'].value_counts().plot(kind='bar')
plt.title('Distribution of Products Across Categories')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.show()

highestCategory = uniqueCategories.idxmax()
highestCategoryCount = uniqueCategories.idxmax()
```

# **IV. Cost and Price Analysis**

```python
averageStandardCost= dataFrame['STANDARD_COST'].mean()
averageListPrice = dataFrame['LIST_PRICE'].mean()

highestStandardCostProduct = dataFrame.loc[dataFrame['STANDARD_COST'].idxmax()] ['PRODUCT_NAME']
highestListPriceProduct = dataFrame.loc[dataFrame['LIST_PRICE'].idxmax()]['PRODUCT_NAME']
```

# **V. Location Analysis**

```python
locationHighestQuantity = dataFrame.groupby('CITY')['QUANTITY'].sum().idxmax()
```

# **VI. Insights and Reporting**

```python
insightsReport = """
Insights and Reporting:
1. Basic Statistics:
  - Mean of Standard Cost: {}
  - Mean of List Price: {}
  - Unique Categories:

{}

2. Category Analysis:
  - Category with Highest Number of Products: {}

3. Cost and Price Analysis:
  - Product with Highest Standard Cost: {}
  - Product with Highest List Price: {}

4. Location Analysis:
  - Location with Highest Quantity of Products: {}

""".format(averageStandardCost, averageListPrice, uniqueCategories, highestCategory, highestStandardCostProduct, highestListPriceProduct, locationHighestQuantity)

print(insightsReport)
```
