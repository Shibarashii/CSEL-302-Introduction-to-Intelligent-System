# I. Import Libraries

```python
#@title Importing Libraries Jr.
import pandas as pd                                   # Data Manipulation
import numpy as np                                    # Mathematical Operations
import matplotlib.pyplot as plt                       # Data Visualization
from sklearn.linear_model import LinearRegression     # Data Modeling
```

# II. Generating Random Data

```python
num_points = 55 #@param {type:"integer"}
X = np.random.randint(0, 100, num_points)
Y = 2 * X + np.random.normal(0, 10, num_points)
```

# III. Creating DataFrame and Saving to CSV

```python
data = pd.DataFrame({'X': X, 'Y': Y})

data.to_csv('data.csv', index=False)

print(data.head())
```

# IV. Visualizing the Data

```python
plt.scatter(data['X'], data['Y'])
xlabel = "Independent Variable (X)" #@param {type:"string"}
plt.xlabel(xlabel)
ylabel = "Dependent Variable (Y)" #@param {type:"string"}
plt.ylabel(ylabel)
plt.title('Scatter Plot of Data')
plt.show()
```

# V. Preparing Data for Modeling

```python
X = data[{'X'}] # Independent Variable
Y = data[{'Y'}] # Dependent Variable
```

# VI. Fitting the Linear Regression

```python
model = LinearRegression()

model.fit(X,Y)
```

# VII. Getting Model Coefficients

```python
slope = model.coef_[0]
intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)
```

# VIII. Visualizing the Linear Regression Line

```python
plt.scatter(X,Y)
plt.plot(X, model.predict(X), color='red')
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.title('Linear Regression')
```

# IX. Making Predicttions

```python
new_X = [[np.random.randint(100, 200)]]

predicted_Y = model.predict(new_X)

print('Value of X: ', new_X)
print('Predicted Value of Y: ', predicted_Y)
```
