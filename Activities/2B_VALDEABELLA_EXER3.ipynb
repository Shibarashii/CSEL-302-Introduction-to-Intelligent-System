```python
#@title I. Data Exploration
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#@title II. Data Processing
dataset = pd.read_csv('/content/Lesson_4_student_grades_dataset.csv')
dataset.head()
X = dataset.drop('Grades', axis=1)
Y = dataset[['Grades']]
```

```python
#@title III. Train-test Split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
model = LinearRegression()
```

```python
#@title IV. Linear Regression
model = LinearRegression()
```

```python
#@title V. Model Training
model.fit(x_train, y_train)
```

```python
#@title VI. Predictions and Evaluations

y_pred = model.predict(x_test)

plt.scatter(y_test, y_pred, color='cyan', label='Predicted Grades')
plt.plot(y_test, y_test, color='red', linestyle='-', label='Actual Grades')
plt.xlabel('Predicted Grades')
plt.ylabel('Actual Grades')
plt.title('Linear Regression Prediction')
plt.show()

print(f"Slope: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

# VII. Feature Importance Analysis

Each feature is crucial to predict the grades of the student

```python
X = dataset[['Hours_of_Study', 'Attendance_Percentage', 'Previous_Exam_Scores', 'Extracurricular_Activities', 'Travel_Time_to_School']]
y = dataset['Grades']

model = LinearRegression()
model.fit(X, y)

coefficients = model.coef_
intercept = model.intercept_
print("Coefficients:", coefficients)
print("Intercept:", intercept)

for column in X.columns:
    plt.figure(facecolor='lightcyan')
    plt.scatter(X[column], y)
    plt.plot(X[column], model.predict(X), color='red', linewidth=5)
    plt.xlabel(f'Independent Variable ({column})', color='black')
    plt.ylabel('Dependent Variable (Grades)', color='black')
    plt.title(f'Linear Regression: {column}', color='black')
    plt.show()

new_data = dataset.sample(1, random_state=42)
new_x = new_data[X.columns]
predicted_y = model.predict(new_x.values.reshape(1, -1))
print("New X:", new_x.values)
print("Predicted Y for New X:", predicted_y[0])
```
